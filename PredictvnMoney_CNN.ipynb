```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.neural_network import MLPClassifier
from google.colab.patches import cv2_imshow
```

```python
mlp = pickle.load(open('/content/moneymodel.sav','rb'))
```

```python
from keras.preprocessing import image
img = cv.imread('/content/50kms.jpeg')
 
img1 = img.copy()
img1 = cv.resize(img1,(200,100))
img1 = image.img_to_array(img1)
img1 = np.expand_dims(img1, axis=0)
img1 =img1.astype(float)/255.0
predict = mlp.predict(img1) 

classes = ['1','10','100','2','20','200','5','50','500']
a = np.argmax(predict)
cv.putText(img, classes[a] , (int(img.shape[1]*10/100),int(img.shape[0]*10/100)), cv.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
cv2_imshow(img)
print(classes[a])

print(predict)
```
